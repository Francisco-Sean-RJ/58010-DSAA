```python
import json

MOVIE_DATABASE = {
    "Inception": {"price": 5.99, "genre": "Action"},
    "The Dark Knight": {"price": 4.99, "genre": "Action"},
    "The Shawshank Redemption": {"price": 6.99, "genre": "Drama"},
    "Pulp Fiction": {"price": 5.49, "genre": "Crime"},
    "The Godfather": {"price": 6.49, "genre": "Crime"},
    "Fight Club": {"price": 4.99, "genre": "Drama"},
    "Forrest Gump": {"price": 5.99, "genre": "Romance"},
    "The Matrix": {"price": 4.49, "genre": "Action"},
    "Schindler's List": {"price": 6.99, "genre": "Drama"},
    "The Lord of the Rings: The Return of the King": {"price": 7.49, "genre": "Adventure"},
}

USERS_FILE = 'users.json'

def load_users():
    try:
        with open(USERS_FILE, 'r') as file:
            return json.load(file)
    except FileNotFoundError:
        return {}

def save_users(users):
    with open(USERS_FILE, 'w') as file:
        json.dump(users, file)

def display_movies(database):
    print("\nMovies Available:")
    for movie, info in database.items():
        print(f"{movie} - {info['genre']} - ${info['price']:.2f}")

def register():
    username = input("Enter your username: ")
    password = input("Enter your password: ")

    users = load_users()

    if username in users:
        print("Username already exists. Please choose a different username.")
    else:
        users[username] = {'password': password, 'movies': []}
        save_users(users)
        print("Registration successful!")

def login():
    username = input("Enter your username: ")
    password = input("Enter your password: ")

    users = load_users()

    if username in users and users[username]['password'] == password:
        print(f"Login successful!")
        return {'username': username, 'movies': users[username]['movies']}
    else:
        print("Invalid username or password. Please try again.")
        return None

def calculate_total_cost(items, database):
    return sum(database[item]["price"] for item in items)

def generate_receipt(items, database, total_cost):
    print("\nReceipt:")
    for item in items:
        print(f"{item} ({database[item]['genre']}) - ${database[item]['price']:.2f}")
    print(f"Total Cost: ${total_cost:.2f}")

# Function to handle movie purchase
def purchase_movies(logged_in_user, database, users):
    display_movies(database)

    selected_movies = input("\nEnter the movies you want to purchase (comma-separated): ").split(',')
    total_cost = calculate_total_cost(selected_movies, database)

    if total_cost > 0:
        print(f"Total cost for movies: ${total_cost:.2f}")
        confirm_purchase = input("Do you want to proceed with the purchase? (yes/no): ").lower()

        if confirm_purchase == 'yes':
            logged_in_user['movies'].extend(selected_movies)
            save_users(users)
            generate_receipt(selected_movies, database, total_cost)
            print("Purchase successful!")
            return True  # Indicate successful purchase
        else:
            print("Purchase canceled.")
    else:
        print("No movies selected.")

    return False  # Indicate unsuccessful purchase

# Welcome message
print("Welcome to TitanMovieShop")

# Main program
logged_in_user = None

while True:
    if logged_in_user:
        print(f"1. List of Movies\n2. Purchase Movies\n3. Exit\nLogged in as {logged_in_user['username']}")
    else:
        print("1. Register\n2. Login\n3. Exit")

    choice = input("Enter your choice (1/2/3): ")

    if choice == '1' and not logged_in_user:
        register()
    elif choice == '2' and not logged_in_user:
        logged_in_user = login()
        if logged_in_user:
            print(f"Welcome, {logged_in_user['username']}!")
    elif choice == '1' and logged_in_user:
        display_movies(MOVIE_DATABASE)
    elif choice == '2' and logged_in_user:
        if purchase_movies(logged_in_user, MOVIE_DATABASE, load_users()):
            break  # Exit the loop after successful purchase
    elif choice == '3':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please choose 1, 2, or 3.")

```
